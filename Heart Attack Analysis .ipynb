```python
data= db.sample(frac=1)
display(data)
```

```python
Y_train = data.loc[:,'target']
X_train = data.drop('target', axis = 1)
Y_train,X_train
```

```python
##KNN 
##random forest classifier
##neural network
```

```python
##random forest classifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
acc_train = []
acc_test = []

#function to call for storing the results
def storeResults(a,b):
 
  acc_train.append(round(a, 3))
  acc_test.append(round(b, 3))
```

```python
x_train, x_test, y_train , y_test = train_test_split(X_train, Y_train, test_size = 0.2, random_state = 12) 
```

```python
##using a max depth of 4.5 for max performance between the training and the validationdata 
##
Classifier_forest= RandomForestClassifier(max_depth = 4.55)
Classifier_forest.fit(x_train,y_train)
```

```python
y_test_forest = Classifier_forest.predict(x_test)
y_train_forest = Classifier_forest.predict(x_train)
```

```python
acc_train_forest = accuracy_score(y_train,y_train_forest)
acc_test_forest = accuracy_score(y_test,y_test_forest)

print("Random forest: Accuracy on training Data: {:.3f}".format(acc_train_forest))
print("Random forest: Accuracy on test Data: {:.3f}".format(acc_test_forest))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
print(classification_report(y_test, y_test_forest))
roc_auc_score(y_test, y_test_forest)
```

```python
##serialize the model with pickle
import pickle
filename = 'Heart_RandomForest_model.pickle'
pickle.dump(Classifier_forest, open(filename, 'wb'))
 
```

```python
##KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
##hyperparameters
##this section are for looking for the better hyperparameter with this model 
leaf_size = list(range(1,50))
n_neighbors =  list(range(1,30))
p =[1,2]
hyperparameters = dict(leaf_size =leaf_size , n_neighbors=n_neighbors , p=p)
knn_classifie2 = KNeighborsClassifier()
#Use GridSearch
clf = GridSearchCV(knn_classifie2, hyperparameters, cv=10)
#Fit the model
best_model = clf.fit(x_train,y_train)
#Print The value of best Hyperparameters
print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
```

```python
y_pred = clf.predict(x_test)
print(classification_report(y_test, y_pred))
roc_auc_score(y_test, y_pred)
```
